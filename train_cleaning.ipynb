## Train data cleaning

```python
import numpy as np
import pandas as pd

train = pd.read_csv("train.csv")
test = pd.read_csv("test_cleaner2.csv")
```

```python
# The train data goes through the same process as test data

sleeps = train["Sleep Duration"]
top_sleep = sleeps.value_counts().index[0]
for sleep in sleeps.value_counts().index:
    sleep_n = len(train[sleep == sleeps])
    if sleep_n < 10:
        train.loc[:, "Sleep Duration"] = train["Sleep Duration"].replace(sleep, top_sleep)
print(train["Sleep Duration"].value_counts())


diet = train["Dietary Habits"]
#print(diet.value_counts())
top_diet = diet.value_counts().index[0]
for d in diet.value_counts().index:
    d_n = len(train[diet == d])
    if d_n < 10:
        train.loc[:, "Dietary Habits"] = train["Dietary Habits"].replace(d, top_diet)
train["Dietary Habits"].value_counts()


cities = train["City"]
top_city = cities.value_counts().index[0]
for city in cities.value_counts().index:
    city_n = len(train[cities == city])
    if city_n < 10:
        train.loc[:, "City"] = train["City"].replace(city, top_city)
train["City"].value_counts()


profs = train["Profession"]
top_prof = profs.value_counts().index[0]
for prof in profs.value_counts().index:
    prof_n = len(train[prof == profs])
    if prof_n < 10:
        print(prof, prof_n)
        train.loc[:, "Profession"] = train["Profession"].replace(prof, top_prof)
print(train["Profession"].value_counts())


degs = train["Degree"]
top_deg = degs.value_counts().index[0]
for deg in degs.value_counts().index:
    deg_n = len(train[deg == degs])
    if deg[0] == "B" or deg == "LLB":
        train.loc[:, "Degree"] = train["Degree"].replace(deg, "Bachelor")
    elif deg[0] == "M" or deg == "LLM":
        train.loc[:, "Degree"] = train["Degree"].replace(deg, "Master")
    if deg_n < 10:
        print(city, city_n)
        train.loc[:, "Degree"] = train["Degree"].replace(deg, top_deg)
print(train["Degree"].value_counts())


train['Pressure'] = train['Academic Pressure'].combine_first(train['Work Pressure'])
position = train.columns.get_loc('Academic Pressure')

columns = (
    train.columns[:position]  # Columns before the position
    .to_list() + ['Pressure'] +  # Add the new column
    train.columns[position + 2:]  # Exclude old columns being replaced
    .to_list()
)
train = train[columns]
train = train.iloc[:, :-1]


train['Satisfaction'] = train['Study Satisfaction'].combine_first(train['Job Satisfaction'])
position = train.columns.get_loc('Study Satisfaction')

columns = (
    train.columns[:position]  
    .to_list() + ['Satisfaction'] + 
    train.columns[position + 2:]  
    .to_list()
)
train = train[columns]
train = train.iloc[:, :-1]
```

```python
train = pd.get_dummies(train, columns=['Gender', 
                                   "Working Professional or Student",
                                   "Dietary Habits",
                                   "Have you ever had suicidal thoughts ?", 
                                   "Family History of Mental Illness",
                                   "Sleep Duration",
                                   "City",
                                   "Profession",
                                   "Degree"], drop_first=True)


dep = train.pop("Depression")
train["y"] = dep
train.head()
```

```python
# Verifying that both train and test data columns match (except "y" column)
traincols = []
testcols = []
for col in train.columns:
    traincols.append(col)
for col in test.columns:
    testcols.append(col)
for col in traincols:
    if col not in testcols:
        print(col + " not in test data")
print("\n")
for col in testcols:
    if col not in traincols:
        print(col + " not in train data")
```
